```python
#step:-2
## identify numerical columns in a dataset
numerical_columns = df.select_dtypes(include=['float64','int64']).columns
print("numerical_columns:",numerical_columns)
```

```python
numerical_columns
```

```python
#step 3:- 
# calculation for mean , median , mode and standard deviation 
for column in numerical_columns:
    print(f"\nColumn:{column}")
    print(f"Mean:{df[column].mean():.2f}")
    print(f"Median:{df[column].median():.2f}")
    print(f"Mode:{df[column].mode()[0]:.2f}")
    print(f"Standard Deviation:{df[column].std():.2f}")
```

```python
#step 4:- 
#loop through each numerical column to plot histograms
import matplotlib.pyplot as plt
for col in numerical_columns:
    plt.hist(df[col].dropna(),bins=10,edgecolor='black',color='hotpink')
    plt.title(f"Histogram for {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
from scipy.stats import skew
for col in numerical_columns:
    col_skewness = skew(df[col].dropna())
    print(f"{col} = skewness:{col_skewness:.2f}")
```

```python
# Check for outliers using the IQR method
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col]<lower_bound) | (df[col]>upper_bound)]
    print(f"{col}:{len(outliers)} outliers defected")
    print(f"{col} Lower Bound :{lower_bound :.2f},Upper Bound :{upper_bound :.2f}")
    print("-" * 50)
```

```python
# boxplot
import seaborn as sns
for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f"Boxplot for {col}")
    plt.xlabel(col)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize numerical columns
scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Compare before and after
print("Before Standardization:")
print(df[numerical_columns].head())

print("After Standardization:")
print(df_standardized[numerical_columns].head())
```

```python
#Conversation of Categorical Data into Dummy Variables
# Specify the categorical columns to encode
categorical_columns = ['Day','City', 'BU', 'Brand', 'Model']
categorical_columns
```

```python
#Bar Chart Analysis for Categorical Column:
for col in categorical_columns:
    plt.figure(figsize=(8,5))
    df[col].value_counts().plot(kind='bar',color='hotpink')
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
#Apply One-hot encoding to the categorical columns
df_encoded = pd.get_dummies(df, columns= categorical_columns)
# Display a portion of the transformed dataset
print(df_encoded.head())
```
