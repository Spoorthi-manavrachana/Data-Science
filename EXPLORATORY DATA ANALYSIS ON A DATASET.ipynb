```python
# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if necessary) - Example: Fill with median values
df.fillna(df.median(),inplace=True)
```

```python
# Check for duplicate rows
print("\n Duplicated rate:",df. duplicated().sum())
```

```python
# Convert data types if necessary (Example: Converting object types to numeric)
df = df.apply(pd.to_numeric,errors = 'coerce')
```

```python
#Step :- 2
# Statistical summary
df.describe()
```

```python
# Compute IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q1 - Q3
print("\n Interquantile Range (IQR):")
print(IQR)
```

```python
# Standard Deviation
print("\n Standard deviation of each columns:")
print(df.std())
```

```python
#Step 3:-
# Histograms for all numerical features
df.hist(figsize=(12,10), bins = 20)
plt.suptitle("Distribution of Numerical Features")
plt.show()
```

```python
# Boxplots for numerical variables
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.suptitle("Boxplot for outlier Detection")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',fmt=".2f",linewidth=0.5 )
plt.suptitle("Correlation Heatmap")
plt.show()
```

```python
# Pairplot for better visualization of relationships
sns.pairplot(df)
plt.show()
```

```python
# 6. Violin Plots (Example: Distribution of ASTV) 
plt.figure(figsize=(8,6))
sns.violinplot(y=df['ASTV'])
plt.title("violin plot of ASTV ", fontsize=14)
plt.show()
```

```python
#Step 4:- 
# Find the most correlated features
correlation_matrics = df.corr()
high_corr_feature = correlation_matrics[abs(correlation_matrics) > 0.5]
print("\nHighly Correlated Features:")
print(correlation_matrics)
```
